```python
portfolio_data_loc = "data/R2500G_proxy_Holdings_clean.csv"
output_path = "data/all_close_prices.csv"

df_holdings = pd.read_csv(portfolio_data_loc)
all_tickers = (
    df_holdings["Ticker"]
    .dropna()
    .astype(str)
    .unique()
    .tolist()
)

price_data = yf.download(
    all_tickers,
    period="1y",
    interval="1d",
    progress=True,
    auto_adjust=True
)

close_df = price_data["Close"]

if isinstance(close_df, pd.Series):
    close_df = close_df.to_frame(name=all_tickers[0])

close_df = close_df.sort_index().ffill()

# 存檔
close_df.to_csv(output_path)
```
